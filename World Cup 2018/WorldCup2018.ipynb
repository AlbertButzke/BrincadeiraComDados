```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "wc2018-players.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "djamshed/fifa-world-cup-2018-players",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

Transformação da base de dados para o pandas

```python
df.rename(columns={'Pos.': 'Position'}, inplace=True)
df
```

```python
df.info()
```

```python
df_age = df #cria um df para mexer na idade
```

```python
from datetime import datetime

def calcular_idade(birth_date_str):
    # Converte a string para data
    birth_date = datetime.strptime(birth_date_str, "%d.%m.%Y")
    # Data de referência: junho de 2018
    ref_date = datetime(2018, 6, 1)
    # Calcula idade
    idade = ref_date.year - birth_date.year - ((ref_date.month, ref_date.day) < (birth_date.month, birth_date.day))
    return idade

df_age['Age'] = df_age['Birth Date'].apply(calcular_idade)
df_age.head()
```
