```python
import math
from typing import List, Dict, Any

def analisar_score(
    query: List[float],
    candidate: Dict[str, Any],
    score_do_console: float
):
    def dot_product(v1, v2):
        return sum(x * y for x, y in zip(v1, v2))
        
    def magnitude(v):
        return math.sqrt(sum(x**2 for x in v))
        
    c_vec = candidate["vetor"]
    
    # 1. Calcula o cosseno puro
    q_mag = magnitude(query)
    c_mag = candidate.get("norma_l2", magnitude(c_vec))
    
    cosseno_puro = dot_product(query, c_vec) / (q_mag * c_mag)
    
    print(f"--- Análise do Candidato: {candidate['document_id']} ---")
    print(f"Cosseno Puro Calculado: {cosseno_puro:.6f}")
    print(f"Score Real do Console:  {score_do_console:.6f}")
    
    # Verifica se o primeiro elemento é negativo para isolar a penalização
    if c_vec[0] < 0:
        print("Atenção: Este candidato possui o primeiro elemento NEGATIVO (vetor[0] < 0).")
        print("O score dele sofre a penalização extra.")
        score_estimado = cosseno_puro / (3*abs(max(query, key=abs)))
        print(f"Score estimado com penalização: {score_estimado}")
    else:
        print("Este candidato NÃO sofre penalização de primeiro elemento negativo.")
        # Se não há penalização negativa, Score = Cosseno / Fator_Query
        # Logo: Fator_Query = Cosseno / Score
        fator_query_estimado = cosseno_puro / score_do_console
        print(f"-> Fator de Normalização da Query estimado para esta consulta: {fator_query_estimado:}")
```

```python
query_vetor = [0.522506, 0.064106, 0.728813, 0.38056, 0.216526]

candidato_exemplo = {
    "document_id": "rh-07",
    "vetor": [0.378142, 0.259474, 0.656852, 0.419534, 0.426871],
    "norma_l2": 1.000000
}
score_console_exemplo = 1.296662

analisar_score(query_vetor, candidato_exemplo, score_console_exemplo)
```
